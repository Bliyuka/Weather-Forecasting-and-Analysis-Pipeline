```python
import requests
from datetime import datetime

# Replace with your actual API key
API_KEY = '78dde17d6fac58c874ed7e93d4a1c294'
CITY = 'Hanoi'
UNITS = 'metric'  # Use 'imperial' for Fahrenheit
# LANG = 'en'       # Language code

# Construct the API URL
url = f'https://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}&units={UNITS}'

# Send the GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    
    # Extract temperature metrics
    temp = data['main'].get('temp')
    feels_like = data['main'].get('feels_like')
    temp_min = data['main'].get('temp_min')
    temp_max = data['main'].get('temp_max')
    
    # Extract atmospheric conditions
    pressure = data['main'].get('pressure')
    humidity = data['main'].get('humidity')
    
    # Extract wind information
    wind_speed = data['wind'].get('speed')
    wind_deg = data['wind'].get('deg')
    wind_gust = data['wind'].get('gust')
    
    # Extract cloud and visibility data
    cloudiness = data['clouds'].get('all')
    visibility = data.get('visibility')
    
    # Extract precipitation data
    rain_1h = data.get('rain', {}).get('1h', 0)
    snow_1h = data.get('snow', {}).get('1h', 0)
    
    # Extract weather description
    weather_main = data['weather'][0].get('main')
    weather_description = data['weather'][0].get('description')
    
    # Extract location and time
    timestamp = data.get('dt')
    timezone_offset = data.get('timezone')
    city_name = data.get('name')
    
    # Convert timestamp to readable format
    local_time = datetime.utcfromtimestamp(timestamp + timezone_offset).strftime('%Y-%m-%d %H:%M:%S')
    
    # Print extracted data
    print(f"Weather data for {city_name} at {local_time}:")
    print(f"Temperature: {temp}°C (Feels like: {feels_like}°C)")
    print(f"Min/Max Temperature: {temp_min}°C / {temp_max}°C")
    print(f"Pressure: {pressure} hPa")
    print(f"Humidity: {humidity}%")
    print(f"Wind: {wind_speed} m/s at {wind_deg}° (Gust: {wind_gust} m/s)")
    print(f"Cloudiness: {cloudiness}%")
    print(f"Visibility: {visibility} meters")
    print(f"Rain (1h): {rain_1h} mm")
    print(f"Snow (1h): {snow_1h} mm")
    print(f"Weather: {weather_main} - {weather_description}")
else:
    print(f"Error fetching data: {response.status_code}")
```
